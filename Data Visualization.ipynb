# Data Visualization
Data visualization is the visual presentation of data or information. The goal of data visualization is to communicate data or information clearly and effectively to readers. Typically, data is visualized in the form of a chart, infographic, diagram or map.

1)Identify trends and outliers

2)Tell a story within the data

3)Reinforce an argument or opinion

4)Highlight an important point in a set of data

## matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python
## seaborn
Seaborn harnesses the power of matplotlib to create beautiful charts in a few lines of code. The key difference is Seaborn's default styles and color palettes, which are designed to be more aesthetically pleasing and modern. Since Seaborn is built on top of matplotlib, you'll need to know matplotlib to tweak Seaborn's defaults.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

#%matplotlib inline sets the backend of matplotlib to the 'inline' backend: 
#With this backend, the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it.
```

```python
days= [1,2,3,4,5,6,7]
max_t=[50,51,49,52,54,46,50]
min_t=[40,41,44,43,47,41,44]
avg_t=[45,46,47,46,48,43,47]
```

```python
# giving labels to x,y axis
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.title('Weather')

#specifing graph lines
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')

#setting legend to best location possible
#The attribute Loc in legend() is used to specify the location of the legend. Default value of loc is loc=”best” (upper left). The strings 'upper left', 'upper right', 'lower left', 'lower right' place the legend at the corresponding corner of the axes/figure
plt.legend(loc='best')
```

```python
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')

#setting location of legend by specifing number
plt.legend(loc=2)
```

```python
x= [1,2,3,4,5,6,7]
y= [1,21,4,43,5,65,7]
#to plot a graph
plt.plot(x,y)
```

```python
#to seek visual properties
# -- stand for type of lines(dashed lines)
# r stand for colour(red)
# D stand for marker(Diamond)
plt.plot(x,y,'--rD')
```

```python
plt.plot(x,y,color='red',marker='^',linestyle= ':')
```

```python
# plot multiple graphs
x1= [4,6,9,12,15,18,21]
y1= [12,58,37,78,23,56,32]
plt.plot(x,y,'--rD', x1,y1,':g*')

x2=[12,23,34,45,56,67,78]
y2=[34,56,25,57,53,37,12]
plt.plot(x,y,'--rD', x1,y1,':g*',x2,y2,'-.y+')
```

```python
company=['GOOGLE','FB','MSFT','AMZN']
revenue=[123,68,92,110]
profit=[50,32,41,10]
```

```python
# taking length of company to make bar as it take both argument as integers
ypos= np.arange(len(company)) #arange function is used to get evenly spaced values within a given interval
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
# Naming x axis bars with company names
plt.xticks(ypos,company)
plt.ylabel('Revenue(bin)')
plt.title('Tech Stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.legend()
```

```python
# ploting two bars
plt.xticks(ypos,company)
plt.ylabel('Revenue(bin)')
plt.title('Tech Stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.bar(ypos,profit,label='Profit')
plt.legend()
```

```python
# to adjust width of two bars
plt.xticks(ypos,company)
plt.ylabel('Revenue(bin)')
plt.title('Tech Stocks')
plt.bar(ypos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(ypos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
# barh() function to make horizontal bars
plt.yticks(ypos,company)

plt.title('Tech Stocks')
plt.barh(ypos-0.2,revenue,label='Revenue')
plt.barh(ypos+0.2,profit,label='Profit')
plt.legend()
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
# by defaul there are 10 bins
plt.hist(blood_sugar, bins=3)
```

```python
# resizing bar size(rwidth() function)
plt.hist(blood_sugar, bins=6, rwidth=0.9)
```

```python
# change colour
plt.hist(blood_sugar, bins=6, rwidth=0.9,color='r')
```

```python
# change type of histogram
plt.hist(blood_sugar, bins=6, rwidth=0.9,histtype='step')
```

```python
bloodsugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
bloodsugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
# two of histogram
plt.title('Blood Sugar Level')
plt.hist([bloodsugar_men,bloodsugar_women], bins=6, rwidth=0.9,color=['green','orange'],label=['Men','Women'])
plt.legend()
```

```python
# changing orientation
plt.title('Blood Sugar Level')
plt.hist([bloodsugar_men,bloodsugar_women], bins=6, rwidth=0.9,color=['green','orange'],label=['Men','Women'],orientation='horizontal')
plt.legend()
```

```python
exp_values=[1400,600,300,410,250]
exp_name=['Home Rent','Food','Phone/Internet Bill','Car','Other Utilities']
```

```python
plt.pie(exp_values,labels=exp_name)
plt.show()
```

```python
# to show axis equally
plt.axis('equal')

plt.pie(exp_values,labels=exp_name)
plt.show()
```

```python
# by default radius of pie is 1 and to change it
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5)
plt.show()
```

```python
# to show percentage of area covered we use autopct argument
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5, autopct='%0.1f%%')
plt.show()
```

```python
# explode argument will take out given peice of pie
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5, autopct='%0.1f%%',explode=[0,0.2,0.1,0,0.1])
plt.show()
```
